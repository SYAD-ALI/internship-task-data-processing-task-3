##Problem Statement 2: Search Query Keyword Analysis

```python
##Problem Statement 2: Search Query Keyword Analysis
#input query
query = "Best laptop deals best price laptop online best"

#convert to lower case
query = query.lower()

import string
query = query.translate(str.maketrans("", "", string.punctuation))

words = query.split()
frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

repeated_keywords = {word: count for word, count in frequency.items() if count > 1}

print(repeated_keywords)
```

##Problem Statement 3: Sensor Data Validation

```python
##Problem Statement 3: Sensor Data Validation
#Input sensor readings

sensor_readings = [3, 4,7,8,10,12,5]
#list to store

valid_readings = []

# Iterate using enumerate to get index (hour) and value
for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:  # Check if even
        valid_readings.append((hour, value))


print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

##Problem Statement 4: Email Domain Usage Analysis

```python
#Problem Statement 4: Email Domain Usage Analysis
# Input list of emails
emails = [
    "ayaan@outlook.com",
    "zoya@gmail.com",
    "faizan@gmail.com",
    "aisha@protonmail.com",
    "imran@gmail.com",
    "sana@outlook.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]   # Extract domain
    domain_count[domain] = domain_count.get(domain, 0) + 1

#Calculate percentage usage
total_emails = len(emails)
print("Email Domain Usage Percentage:")

for domain, count in domain_count.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

##Problem Statement 5: Sales Spike Detection

```python
#Problem Statement 5: Sales Spike Detection
# Input sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]
average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.3
print("Sales Spike Days:")

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

##Problem Statement 6: Duplicate User ID Detection

```python
#Problem Statement 6: Duplicate User ID Detection
# Input user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
# Count occurrences
id_count = {}

for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Identify duplicates
print("Duplicate User IDs:")

for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
